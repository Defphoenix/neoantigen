```python
def find_difference_range(WT, MT):
    if WT==None or MT==None:
        return None,None
    len1, len2 = len(WT), len(MT)
    min_length = min(len1, len2)
    if len1 == len2:
        # 找到第一个不同的字符位置
        start = 0
        while start < min_length and WT[start] == MT[start]:
            start += 1
        # 如果所有字符都相同，检查长度是否不同
        if start == min_length:
            if len1 == len2:
                return None  # 序列完全相同
            else:
                return start, max(len1, len2) - 1  # 序列长度不同，从 start 到末尾不同
        
        # 找到最后一个不同的字符位置
        end1, end2 = len1 - 1, len2 - 1
        while end1 >= 0 and end2 >= 0 and WT[end1] == MT[end2]:
            end1 -= 1
            end2 -= 1
        end = max(end1, end2)
        return start, end
    elif len1 <= len2:
        start = 0
        while start < min_length and WT[start] == MT[start]:
            start += 1
        # 如果所有字符都相同，检查长度是否不同
        return start, start+len2-len1  # 序列长度不同，从 start 到末尾不同
    elif len1 >= len2:
        start = 0
        while start < min_length and WT[start] == MT[start]:
            start += 1
        return start,start+1
```
